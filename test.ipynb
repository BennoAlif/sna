```python
tweets_data = []
tweets_user = []

for response in tweepy.Paginator(client.search_recent_tweets,
                                query="#racunshopee lang:id",
                                start_time=start_time,
                                end_time=end_time,
                                tweet_fields = ["created_at", "text", "source"],
                                user_fields = ["name", "username", "location", "verified", "description"],
                                max_results = 100,
                                expansions='author_id', limit=10):
  
    tweets_data += response.data
    tweets_user += response.includes["users"]
```

```python
dict(tweets_data[0])
```

```python
results = []

for tweet, user in zip(tweets_data, tweets_user):
  tweet_info = {
    'id': tweet.id,
    'created_at': tweet.created_at,
    'text': tweet.text,
    'source': tweet.source,
    'name': user.name,
    'username': user.username,
    'location': user.location,
    # 'verified': user.verified,
    'description': user.description
  }
  results.append(tweet_info)


results_df = pd.DataFrame(results)

results_df.info()

# pd.DataFrame(tweets_data)["text"]
```
